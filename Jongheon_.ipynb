```python
#출근 시간 역별 승차 인원
cursor = subway.aggregate([
    {
        '$match': {
            'use_mon': '202104'
        }
    }, {
        '$group': {
            '_id': '$sub_sta_nm', 
            'total6': {
                '$sum': '$06시-07시 승차인원'
            }, 
            'total7': {
                '$sum': '$07시-08시 승차인원'
            }, 
            'total8': {
                '$sum': '$08시-09시 승차인원'
            }
        }
    }, {
        '$addFields': {
            '06시~09시 승차인원': {
                '$add': [
                    '$total6', '$total7', '$total8'
                ]
            }
        }
    }, {
        '$project': {
            '06시~09시 승차인원': 1
        }
    }, {
        '$sort': {
            '06시~09시 승차인원': -1
        }
    }, {
        '$limit': 20
    }
])

people_by_station = pd.DataFrame(list(cursor))

fig = px.bar(people_by_station,
              x = "_id",
              y = "06시~09시 승차인원",
              color = "_id"
              )
fig.show()
```
